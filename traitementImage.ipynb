## <span style='color:dark green'> "Manipuler" les pixels d'une image </code>
Une image numérique matricielle a pour structure un tableau de pixels

Chaque pixel a une position précise et une couleur précise.
Les positions sont repérées à l'aide de deux coordonnées x et y dont l'origine se situe en haut à gauche de l'image. La première position est <b>(0,0)</b>
* Il faudra donc une boucle <code> for x in range(largeur) </code> pour parcourir les pixels sur toute la largeur
* Il faudra donc une boucle <code> for y in range(hauteur) </code> pour parcourir les pixels sur toute la hauteur




Il existe différentes façon de coder la couleur d’un pixel. La plus commune est sans doute le RGB (Red, Green, Blue). Chaque couleur est décrite en fonction de son intensité (saturation) en rouge, en vert et en bleu. Chacune de ces saturations est codée sur 8 bits ce qui laisse 2<sup>8</sup> = 256 valeurs de saturation possibles. Chaque <b>composante</b> est donc codée par une <b>valeur comprise entre 0 et 255</b>. Une « couleur » est donc codée sur 24 bits soit 3 octets.

Chaque composante est codée par une valeur comprise entre 0 et 255

* la fonction <code>im.putpixel((x,y), ((intensiteRed,intensiteGreen,intensiteBlue))))</code> permet de fixer une couleur (RGV) au pixel situé en (x,y).

### Inversion des couleurs

Dans le programme qui suit, nous allons répéter pour chque pixel le traitement suivant :
* lire la valeur de la couleur du pixel situé à laposition (x,y)  : getpixel
* remplacer le rouge par le vert
* remplacer le vert par le bleu
* remplacer le bleu par le rouge
* ecrire la nouvelle couleur du pixel : putpixel
 
 <img src="images/101.jpg"      alt="affiche des 101" width="100">
         

<b><span style='color:purple>'> Activité 9 / Etudier le programme suivant et rechercher la ligne où se trouve l'inversion de couleurs. Tester ensuite le programme. </span></b>

```python
from PIL import Image
image = Image.open("Images/101.jpg") 
largeur, hauteur = image.size
for y in range(hauteur): 
    for x in range(largeur):
        rouge,vert,bleu=image.getpixel((x,y))
        image.putpixel((x,y),(vert,bleu,rouge))
image
```

### Modification des couleurs de l'affiche de la série "13 reasons why"
 <img src="images/13.jpg"      alt="affiche de 13 reasons why" width="100">
Il est possible de modifier les couleurs de façon arbitraire. par exemple si je fais : <code> nouveau_rouge = (rouge + 56) </code> alors j'accentue l'intensité du rouge. La valeur <b>+56</b> est choisie au hasard.
Comme la valeur ne doit pas dépasser la valeur 255 alors on peut effectuer l'un des traitement suivant  :
* <code>  nouveau_rouge = (rouge + 56) %255 </code>

ou

* <code>  nouveau_rouge = (rouge + 56) <br/>  if nouveau_rouge >255: nouveau_rouge=255 </code>

<b><span style='color:purple>'> Activité 10 / Tester le programme suivant :    </span></b>

```python
from PIL import Image

image = Image.open("images/13.jpg")

largeur, hauteur = image.size

for y in range(hauteur):
    for x in range(largeur):
        rouge,vert,bleu = image.getpixel((x,y))
       
        nouveau_rouge = (rouge + 56) %255 #modification de la composante rouge par une valeur 
        nouveau_vert = (vert + 200) %255 #modification de la composante verte
        nouveau_bleu = (bleu  - 45) %255 #modification de la composante bleue
       
        image.putpixel((x,y), (nouveau_rouge,nouveau_vert,nouveau_bleu))  # ecriture de la couleur du pixel
        
image
```
